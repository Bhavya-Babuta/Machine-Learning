```python
Y=train['target']
Y=np.array(Y).reshape(Y.shape[0],1)
X=train.drop('target',axis=1)
```

```python
X.head(20)
```

```python
X = (X - np.min(X)) / (np.max(X) - np.min(X)).values
```

```python
X.head(20)
```

```python
def initialize_parameters(dimension):
    weight = np.full((dimension,1),0)
    bias = 0.0
    return weight,bias
```

```python
def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))

def forward_prog(X,Y,W,b):
    yhat= sigmoid(X.dot(W)+b)
    loss = -(Y*np.log(yhat) + (1-Y)*np.log(1-yhat))
    cost = np.sum(loss) / X.shape[1]
    return yhat,loss,cost

def backward_prog(X,Y,predictions,W,b,learning_rate):    
    W=W-(learning_rate*(X.T.dot(predictions-Y)/X.shape[1]))
    b=b-(learning_rate*((predictions-Y).sum()/X.shape[1]))
    return W,b   

def logisitic_regression(X,Y,learning_rate,iterations):
    cost=[]
    loss=[]
    W,b=initialize_parameters(X.shape[1])
    for i in range(1,iterations+1):
        predicitions,loss_i,cost_i=forward_prog(X,Y,W,b)
        W,b=backward_prog(X,Y,predicitions,W,b,learning_rate)
        cost.append(cost_i)
        loss.append(loss_i)
    return W,b,cost,loss

def classification_rate(Y,P):
    return np.mean(Y == P)

def predict(W,b,X):
    z = sigmoid(X.dot(W) + b)           
    y_prediction = np.round(z)
    return y_prediction; 
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size = 0.2,random_state=0)
W,b,cost,loss=logisitic_regression(x_train,y_train,0.001,1000)
```

```python
pred=predict(W,b,x_test)
print(classification_rate(y_test,pred))
plt.figure()
sns.lineplot(x=list(range(1,1001)),y=np.array(cost)[:,0])
```

```python
from sklearn.metrics import confusion_matrix
cm_lr = confusion_matrix(y_test,pred)
sns.heatmap(cm_lr,annot=True)
```
